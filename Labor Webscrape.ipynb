```python
states = ['Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Deleware','Florida','Georgia',
          'Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Maryland',
          'Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey',
          'New Mexico','New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina',
          'South Dakota','Tennessee','Texas','Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming']
```

```python
# list to store each df for concatenation

dfObj = []
```

```python
# loop and scrape each states unemployment information

for x in range(len(ids)):
    url = 'https://data.bls.gov/timeseries/'+ids[x]
    tables = pd.read_html(url)
    df = tables[1]
    df = df.iloc[:122]
    df['State'] = states[x]
    
    df.columns = ['Year', 'Month', 'Labor Force Participation Rate', 'Employment-Population Ratio', 'Labor Force', 'Employment', 'Unemployment', 'Unemployment Rate', 'State']
    print(df)
    dfObj.append(df)
```

```python
# concatenate the 50 dataframes into a single dataframe
df = pd.concat(dfObj)
result = df[['State','Year', 'Month', 'Labor Force Participation Rate', 'Employment-Population Ratio', 'Labor Force', 'Employment', 'Unemployment', 'Unemployment Rate']]
```

```python
#turn data into strings and remove the string values

result['Labor Force Participation Rate'] = result['Labor Force Participation Rate'].str.replace('R','')
result['Labor Force Participation Rate'] = result['Labor Force Participation Rate'].str.replace('(','')
result['Labor Force Participation Rate'] = result['Labor Force Participation Rate'].str.replace(')','')
result['Labor Force Participation Rate'] = result['Labor Force Participation Rate'].str.replace('P','')

result['Labor Force'] = result['Labor Force'].str.replace('R','')
result['Labor Force'] = result['Labor Force'].str.replace('(','')
result['Labor Force'] = result['Labor Force'].str.replace(')','')
result['Labor Force'] = result['Labor Force'].str.replace('P','')

result['Employment-Population Ratio'] = result['Employment-Population Ratio'].str.replace('R','')
result['Employment-Population Ratio'] = result['Employment-Population Ratio'].str.replace('(','')
result['Employment-Population Ratio'] = result['Employment-Population Ratio'].str.replace(')','')
result['Employment-Population Ratio'] = result['Employment-Population Ratio'].str.replace('P','')

result['Employment'] = result['Employment'].str.replace('R','')
result['Employment'] = result['Employment'].str.replace('(','')
result['Employment'] = result['Employment'].str.replace(')','')
result['Employment'] = result['Employment'].str.replace('P','')

result['Unemployment'] = result['Unemployment'].str.replace('R','')
result['Unemployment'] = result['Unemployment'].str.replace('(','')
result['Unemployment'] = result['Unemployment'].str.replace(')','')
result['Unemployment'] = result['Unemployment'].str.replace('P','')

result['Unemployment Rate'] = result['Unemployment Rate'].str.replace('R','')
result['Unemployment Rate'] = result['Unemployment Rate'].str.replace('(','')
result['Unemployment Rate'] = result['Unemployment Rate'].str.replace(')','')
result['Unemployment Rate'] = result['Unemployment Rate'].str.replace('P','')
```

```python
# turn data back into integers

result['Labor Force Participation Rate'] = pd.to_numeric(result['Labor Force Participation Rate'])
result['Labor Force'] = pd.to_numeric(result['Labor Force'])
result['Employment-Population Ratio'] = pd.to_numeric(result['Employment-Population Ratio'])
result['Employment'] = pd.to_numeric(result['Employment'])
result['Unemployment'] = pd.to_numeric(result['Unemployment'])
result['Unemployment Rate'] = pd.to_numeric(result['Unemployment Rate'])

result
```
